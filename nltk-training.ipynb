NLTK BASICS IN ACTION
- PUNCTUATION REMOVAL

```python
from nltk.stem import porter
from nltk.corpus import stopwords
import string

stp_wrds = set(stopwords.words('english'))
pd.set_option('display.max_colwidth', 100)

#remove punctuation
def remove_punctuation(elem):
    text = ''.join([char for char in elem if char not in string.punctuation])
    return text

dataset['punctuation_removed'] = dataset['body_text'].apply(lambda x: remove_punctuation(x))
dataset.head()
```

- TOKENIZATION

```python
#tokenization
def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

dataset['tokenized'] = dataset['body_text_clean'].apply(lambda x: tokenize(x))
dataset.head()
```

STOPWORD REMOVAL

```python
#remove stop_words
def remove_stop_words(tokenized_list):
    text = [word for word in tokenized_list if word not in stopwords.words('english')]
    return text

dataset['stopwrd_removed'] = dataset['tokenized'].apply(lambda x: remove_stop_words(x))
dataset.head()
```

- STEMMING
(Returning a word to it's root or base form')

```python
#stemming
from nltk import PorterStemmer

ps = PorterStemmer()
for i in ['grow', 'grows', 'growing']:
    print(ps.stem(i))
```

```python
def stemming(tokenized_list):
    text = [ps.stem(word) for word in tokenized_list]
    return text

dataset['stemmed'] = dataset['stopwrd_removed'].apply(lambda x: stemming(x))
dataset.head()
```

- LEMMATIZATION (Returning a synonym of a word)

```python
#lemmatization
from nltk import WordNetLemmatizer

ls = WordNetLemmatizer()

def lemmatizing(tokenized_list):
    text = [ls.lemmatize(word) for word in tokenized_list]
    return text

dataset['lemmatized'] = dataset['stemmed'].apply(lambda x: lemmatizing(x))
dataset.head()
```
